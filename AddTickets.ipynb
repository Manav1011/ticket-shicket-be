```python
event_id='b7e2a1c2-3f4d-4e6a-9b8c-1a2b3c4d5e6f'
event_day_id='2024-06-01'
```

```python
# 100 tickets
ticket_count = 100
```

```python
import redis

# Connect to Redis (default localhost:6379)
r = redis.Redis(host='localhost', port=63799, db=0)
```

```python
secret = "your_jwt_secret_here"
```

```python
# create payload for tickets
# {
#     "event_id"
#     "event_day_id"
#     "ticket_id"
#     "ticket_index"
# }

import jwt
bitmap_key = f'{event_id}:{event_day_id}:bitmap'
```

```python


for i in range(ticket_count):
    r.setbit(bitmap_key, i, 0)

for i in range(ticket_count):
    ticket_id = f'ticket-{i+1:03d}'
    payload = {
        "event_id": event_id,
        "event_day_id": event_day_id,
        "ticket_id": ticket_id,
        "ticket_index": i
    }
    token = jwt.encode(payload, secret, algorithm="HS256")    
    print(token)
```

```python
bitmap_bytes = r.get(bitmap_key)



if bitmap_bytes is not None:
    # Convert bytes to a list of bits
    bits = []
    for byte in bitmap_bytes:
        for i in range(8):
            bits.append((byte >> (7 - i)) & 1)
    # Print only the relevant ticket_count bits
    print(bits[:ticket_count])
    print(len(bits[:ticket_count]))
else:
    print("Bitmap not found in Redis.")
```
