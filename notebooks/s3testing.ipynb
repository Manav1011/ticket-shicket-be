```python
import boto3
from botocore.exceptions import ClientError
```

```python
S3_ENDPOINT = "http://localhost:4566"
BUCKET_NAME = "my-bucket"
FILE_NAME = "/home/manav1011/Documents/ticket-shicket-be/CLAUDE.md"
DOWNLOADED_FILE = "downloaded_test.txt"
```

```python
def get_s3_client():
    return boto3.client(
        "s3",
        endpoint_url=S3_ENDPOINT,
        aws_access_key_id="test",
        aws_secret_access_key="test",
        region_name="us-east-1",
    )
    
def create_bucket(s3):
    try:
        s3.create_bucket(Bucket=BUCKET_NAME)
        print(f"✅ Bucket '{BUCKET_NAME}' created")
    except ClientError as e:
        if e.response["Error"]["Code"] == "BucketAlreadyOwnedByYou":
            print(f"ℹ️ Bucket '{BUCKET_NAME}' already exists")
        else:
            raise


def upload_file(s3):
    with open(FILE_NAME, "w") as f:
        f.write("Hello from LocalStack S3!")

    s3.upload_file(FILE_NAME, BUCKET_NAME, FILE_NAME)
    print(f"✅ Uploaded '{FILE_NAME}'")


def list_files(s3):
    response = s3.list_objects_v2(Bucket=BUCKET_NAME)
    print("📂 Files in bucket:")

    if "Contents" in response:
        for obj in response["Contents"]:
            print(f" - {obj['Key']}")
    else:
        print(" (empty)")


def download_file(s3):
    s3.download_file(BUCKET_NAME, FILE_NAME, DOWNLOADED_FILE)
    print(f"✅ Downloaded as '{DOWNLOADED_FILE}'")

def generate_presigned_url(s3):
    url = s3.generate_presigned_url(
        ClientMethod="get_object",
        Params={
            "Bucket": "my-bucket",
            "Key": FILE_NAME,
        },
        ExpiresIn=3600,  # 1 hour
    )

    print("🔗 Presigned URL:")
    print(url)
    
s3 = get_s3_client()

# create_bucket(s3)
# upload_file(s3)
# list_files(s3)
# download_file(s3)
generate_presigned_url(s3)
```
